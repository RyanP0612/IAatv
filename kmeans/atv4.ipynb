```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


data = {'Substância': ['Álcool', 'Gasolina', 'Leite', 'Querosene', 'Óleo', 'Vinho'],
        'Concentração (%)': [12.5, 0.1, 4.0, 1.2, 0.5, 15.0],
        'Teor Alcoólico (%)': [50, 0.05, 0.01, 0.02, 0.01, 12.5]}

df = pd.DataFrame(data)


X = df[['Concentração (%)', 'Teor Alcoólico (%)']].values


kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)


print(df)


plt.scatter(df['Concentração (%)'], df['Teor Alcoólico (%)'], c=df['Cluster'], cmap='viridis')
plt.xlabel('Concentração (%)')
plt.ylabel('Teor Alcoólico (%)')
plt.title('Agrupamento de Substâncias')
plt.show()
```
