```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


data = {'Setores': [1, 2, 3, 4, 5, 6, 7, 8],
        'Nº de Produtos Fabricados': [100, 50, 15, 200, 500, 1000, 375, 450]}

df = pd.DataFrame(data)

X = df['Nº de Produtos Fabricados'].values.reshape(-1, 1)


kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)


print(df)


plt.scatter(df['Setores'], df['Nº de Produtos Fabricados'], c=df['Cluster'], cmap='viridis')
plt.xlabel('Setores')
plt.ylabel('Nº de Produtos Fabricados')
plt.title('Agrupamento de Setores Mais Produtivos')
plt.show()
```
