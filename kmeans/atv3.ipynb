```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


data = {'Dia da Semana': [1, 2, 3, 4, 5, 6, 7],
        'Nº de Clientes': [15, 20, 30, 100, 350, 500, 700]}

df = pd.DataFrame(data)


X = df['Nº de Clientes'].values.reshape(-1, 1)


kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)


print(df)

plt.scatter(df['Dia da Semana'], df['Nº de Clientes'], c=df['Cluster'], cmap='viridis')
plt.xlabel('Dia da Semana')
plt.ylabel('Nº de Clientes')
plt.title('Agrupamento de Dias de Maior Movimento')
plt.show()
```
