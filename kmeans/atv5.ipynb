```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
# Criar DataFrame com os dados das máquinas
data = {
    'Máquina': ['M1', 'M2', 'M3', 'M4', 'M5', 'M6', 'M7', 'M8', 'M9', 'M10'],
    'Temperatura (°C)': [70.2, 65.1, 75.5, 80.3, 68.7, 72.9, 78.6, 66.4, 73.1, 69.5],
    'Vibração': [12.5, 8.2, 15.6, 10.2, 11.8, 14.3, 9.8, 8.9, 13.7, 12.1],
    'Corrente': [4.7, 3.9, 5.1, 4.5, 4.2, 5.3, 4.8, 4.0, 5.0, 4.3]
}

df = pd.DataFrame(data)

# Criar uma matriz numpy
X = df[['Temperatura (°C)', 'Vibração', 'Corrente']].values

# Aplicar K-means com k=3 (por exemplo)
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)



# Plotar o boxplot utilizando Seaborn
plt.figure(figsize=(10, 6))
sns.boxplot(x='Cluster', y='Temperatura (°C)', data=df)
plt.title('Boxplot da Temperatura por Cluster')
plt.show()
```
